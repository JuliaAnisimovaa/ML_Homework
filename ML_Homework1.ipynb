#0. Описать задачу словами

Набор данных взят с Kaggle: https://www.kaggle.com/datasets/teejmahal20/airline-passenger-satisfaction/?select=test.csv

Набор данных состоит из 103904 и 25976 записей в обучающем и тестовом режимах соответственно. Данный набор данных содержит опрос об удовлетворенности авиапассажиров. Он состоит из 25 столбцов, таких как «Возраст», «Пол», «Класс путешествия», «Задержки прибытия и отправления», а также функций, влияющих на уровень удовлетворенности клиентов, таких как «Обслуживание на борту», «Чистота», «Комфорт сиденья», «Обработка багажа» и т.д. Набор данных состоит из столбца или функция под названием «удовлетворение» , которая описывает общий уровень удовлетворенности клиента. Он имеет два значения: «нейтральный или неудовлетворенный» и «удовлетворенный». Этот признак удовлетворенности считается признаком маркировки, поскольку он передает общее впечатление клиента на основе оценок.

Таким образом, поставлена задача <b>классификации</b>: необходимо предсказать к какому из двух уровней удовлетворённости относится пассажир: «<b>нейтральный</b> или <b>неудовлетворенный</b>» и «<b>удовлетворенный</b>».

Признаки.

<b> 1. Gender </b>- Пол. Бинарный признак. Типы: Female (Женский), Male (Мужской).

<b> 2. Customer Type</b>  - Тип клиента. Бинарный признак. Типы: Loyal customer (Лояльный клиент) и Disloyal customer (Нелояльный клиент).
Лояльный клиент - это клиент, лояльный к авиакомпании, то есть чаще предпочитающий пользоваться услугами данной авиакомпании нежели другой. Нелояльный - чаще пользующийся услугами другой компании.

<b> 3. Age</b> - Возраст. Количественный признак.
<b> 4. Type of Travel </b>- Цель полета пассажира. Бинарный признак. Типы: Personal Travel (Личная поездка), Business Travel (Деловая поездка).

<b> 5. Class</b> - Туристический класс, которым летит пассажир. Категориальный признак. Типы: Business (Бизнес), Eco (Эконом), Eco Plus (Эконом Плюс).

<b> 6. Flight Distance </b> - Расстояние полета. Количественный признак.

<b> 7. Inflight wifi service</b> - уровень удовлетворенности услугой Wi-Fi - от 1 до 5. Количественный признак.

<b> 8. Departure/Arrival time convenient</b> - Уровень удовлетворенности тем, насколько удобно было время отправления/прибытия. Количественный признак.

<b> 9. Ease of Online booking</b> - Уровень удовлетворенности простотой онлайн-бронирования. Количественный признак.

<b> 10. Gate location</b> - Уровень удовлетворенности расположением выходов. Количественный признак.

<b> 11. Food and drink </b>- Уровень удовлетворенности едой и напитками. Количественный признак.

<b> 12. Online boarding</b> - Уровень удовлетворенности онлайн регистрацией на рейс. Количественный признак.

<b> 13. Seat comfort </b>- Уровень удовлетворенности комфортабельностью кресел. Количественный признак.

<b> 14. Inflight entertainment</b> - Уровень удовлетворенности развлечениями во время полета. Количественный признак.

<b> 15. On-board service</b> - Уровень удовлетворенности услугами на борту. Количественный признак.

<b> 16. Leg room service</b> - Уровень удовлетворенности количеством места для ног. Количественный признак.

<b> 17. Baggage handling</b> - Уровень удовлетворенности услугами по транспортировке багажа. Количественный признак.

<b> 18. Checkin service</b> - Уровень удовлетворенности услугой регистрации на рейс. Количественный признак.

<b> 19. Inflight service</b> - Уровень удовлетворенности услугами во время полета. Количественный признак.

<b> 20. Cleanliness</b> - Уровень удовлетворенности чистотой. Количественный признак.

<b> 21. Departure Delay in Minutes </b> - Задержка отправления в минутах. Количественный признак.

<b> 22. Arrival Delay in Minutes</b> - Задержка прибытия в минутах. Количественный признак.

<b> 23. Satisfaction</b> - Уровень удовлетворенности. Бинарный признак. Типы: satisfaction, neutral or dissatisfied.

#4. Обработать категориальные признаки

```python
categ_columns = [c for c in data_raw.columns if data_raw[c].dtype.name == 'category']
categ_columns
```

Рассмотрим бинарные признаки:

```python
categorical_columns = [c for c in data_raw.columns if data_raw[c].dtype.name == 'category']; describe = data_raw.describe(include=['category'])
```

```python
for col in categ_columns:
    if describe[col]['unique'] == 2:
        data_raw[col] = data_raw[col].astype('category')
        data_raw[col] = data_raw[col].cat.codes
```

Рассмотри небинарные признаки:

```python
nonbinary_columns = [c for c in categorical_columns if data_describe[c]['unique'] > 2]
nonbinary = pd.get_dummies(data_raw[nonbinary_columns])
```

#5. Провести нормализацию

Некоторые алгоритмы машинного обучения чувствительны к масштабированию данных,например, метод ближайших соседей, метод опорных векторов и т.д.

Из прошлых пунктов, исходя из матрицы корреляции видно, что между признаками "Departure Delay in Minutes" и "Arrival Delay in Minutes" высокая корреляция. Можно удалить один признак из пары сильно коррелирующих признаков для того, чтобы уменьшить число рассматриваемых признаков. То есть можно удалить признак Arrival Delay in Minutes.

```python
data_raw.drop("Arrival Delay in Minutes", axis =1, inplace= True)
```

```python
num_columns = [c for c in data_raw.columns if data_raw[c].dtype.name != 'category' and c != 'satisfaction' ]
data_numerical = data_raw[num_columns]
```

```python
data_raw[num_columns] = (data_raw[num_columns] - data_raw[num_columns].mean(axis = 0))/data_raw[num_columns].std(axis = 0)
data_raw.describe()
```

Соберём все данные:

```python
Y= data_raw['satisfaction']
```

```python
binary_columns = [c for c in categorical_columns if data_describe[c]['unique'] == 2]
data_raw = pd.concat((data_raw[num_columns], nonbinary, data_raw[binary_columns]), axis = 1)
```

#6. Разбить данные на обучающую и тестовую выборки

```python
X = data_raw
```

```python
X.columns
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 555)

N_train, _ = X_train.shape
N_test,  _ = X_test.shape
```

```python
print(N_train, N_test)
```

#7. Запустить классификатор (регрессию) ближайших соседей

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train, y_train)
```

```python
search = GridSearchCV(KNeighborsClassifier(), param_grid={'n_neighbors': range(1, 21)}, cv=5, n_jobs=-1).fit(X_train, y_train)

print(1 - search.best_score_)
print( search.best_estimator_.n_neighbors)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report

kNNf = KNeighborsClassifier()
kNNf.n_neighbors = search.best_estimator_.n_neighbors
kNNf.fit(X_train, y_train)

y_train_prediction = kNNf.predict(X_train)
y_test_prediction = kNNf.predict(X_test)
cm = confusion_matrix(y_train, y_train_prediction)
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=kNNf.classes_)
disp.plot()
```

```python
print(classification_report(y_test, y_test_prediction))
```

```python
err_train = np.mean(y_train != kNNf.predict(X_train))
err_test  = np.mean(y_test  != kNNf.predict(X_test))
```

```python
print(err_train, err_test)
```

```python
from sklearn import ensemble
rf = ensemble.RandomForestClassifier(n_estimators = 1000)
rf.fit(X_train, y_train)

err_train = np.mean(y_train != rf.predict(X_train))
err_test  = np.mean(y_test  != rf.predict(X_test))

print(err_train, err_test)
```

<b>Вывод:</b> Безусловно, алгоритм Random Forest лучше справляется с задачей, чем метод ближайших соседей (KNN), так как его ошибка на тестовой выборке ниже.

Ошибки: 0.0 на обучающей выборке и 0.0357 на тестовой выборке для Random Forest; для KNN — 0.048 на обучающей и 0.060 на тестовой.
